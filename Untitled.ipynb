```python
corpus = """low
lower
lowest
new
newer
widest"""

with open("corpus.txt", "w") as file:
    file.write(corpus)

print("corpus.txt created successfully!")
```

```python
with open("corpus.txt", "r") as file:
    corpus = file.read().splitlines()

print(corpus)
```

```python
corpus = [word.lower().strip() for word in corpus]

print(corpus)
```

```python
vocab = {}
for word in corpus:
    chars = " ".join(word) + " </w>"
    vocab[chars] = vocab.get(chars, 0) + 1
print(vocab)
```

```python
from collections import defaultdict
def get_pair_frequencies(vocab):
    pairs = defaultdict(int)
    for word, freq in vocab.items():
        symbols = word.split()
        for i in range(len(symbols) - 1):
            pairs[(symbols[i], symbols[i + 1])] += freq
    return pairs
pairs = get_pair_frequencies(vocab)
for pair, freq in pairs.items():
    print(pair, ":", freq)
```

```python
best_pair = max(pairs, key=pairs.get)
print("Most Frequent Pair:", best_pair)
print("Frequency:", pairs[best_pair])
```

```python
def merge_pair(pair, vocab):
    merged_vocab = {}
    bigram = " ".join(pair)
    replacement = "".join(pair)
    for word in vocab:
        new_word = word.replace(bigram, replacement)
        merged_vocab[new_word] = vocab[word]
    return merged_vocab
vocab = merge_pair(best_pair, vocab)
print(vocab)
```

```python
for i in range(5):
    pairs = get_pair_frequencies(vocab)
    best_pair = max(pairs, key=pairs.get)
    print(f"Merge {i+1}: {best_pair}")
    vocab = merge_pair(best_pair, vocab)
print("\nFinal Vocabulary:")
for word in vocab:
    print(word)
```
